# Project Title

### Banking Transaction Analysis Using NumPy

### Problem Statement :-

Banks process thousands of transactions every day. Analyzing transaction data manually is time-consuming and inefficient. This project provides a simple way to analyze daily transaction data and gain useful insights into banking activity.

### Objectives :- 
- Store daily transaction data in a 1D NumPy array.
- Calculate total and average transactions.
- Identify the highest and lowest transaction days.
- Measure transaction variation using transaction range.
- Find the median transaction value.
- Determine the number of above-average transaction days.

### Solution :- 

```python
import numpy as np

transactions = np.array([1200, 1500, 1800, 1100, 2200, 1900, 1400])
```

### Banking Analysis

```python
# Calculate total transactions
total_transactions = np.sum(transactions)

# Calculate average transactions
average_transactions = np.mean(transactions)

# Find highest transaction value
highest_transaction = np.max(transactions)

# Find lowest transaction value
lowest_transaction = np.min(transactions)

# Count days with transactions above average
above_average_days = np.sum(transactions > average_transactions)

# Difference between highest and lowest transactions
transaction_range = highest_transaction - lowest_transaction

# Middle transaction value after sorting the data
median_transactions = np.median(transactions)
```

```python
print("Total Transactions:", total_transactions)
print("Average Transactions:", average_transactions)
print("Highest Transaction:", highest_transaction)
print("Lowest Transaction:", lowest_transaction)
print("Transaction Range:", transaction_range)
print("Median Transactions:", median_transactions)
print("Days Above Average:", above_average_days)
```

### Business Impact :- 
- Helps monitor banking activity.
- Identifies peak and low transaction days.
- Supports resource and cash-flow planning.
- Helps detect unusual transaction patterns.
- Reduces manual analysis effort.
- Enables data-driven decision making.

### Conclusion :-

This project uses NumPy to analyze daily banking transactions efficiently. By calculating key metrics such as total transactions, average transactions, median transactions, transaction range, and above-average transaction days, banks can better understand transaction patterns and improve operational planning.
